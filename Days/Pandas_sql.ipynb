### **--Find Top  Reasons for returning.**

```python
df.columns
```

```python
k = pd.read_sql('''


SELECT return_reason, Count(return_reason) FROM store
GROUP BY return_reason
having Return_Reason != 'Not Returned'
ORDER BY Count(return_reason) DESC


''', conn)
k
```

```python
import matplotlib.pyplot as plt
```

```python
k.plot(kind='bar', x='Return_Reason', y='Count(return_reason)',color = 'red')
```

### **Check whether new customers are returning higher or not.**

```python
df.columns
```

```python
df['Customer_Duration'].value_counts()
```

```python
pd.read_sql('''


SELECT Customer_Duration, Count(Customer_Duration) FROM store
where Return_Reason != 'Not Returned'
GROUP BY Customer_Duration

''', conn)
```

### **Find Top 20 cities and states having higher return.**

```python
pd.read_sql('''


SELECT States,city, Count(Customer_Duration) FROM store
where Return_Reason != 'Not Returned'
GROUP BY States,city
order by Count(Customer_Duration) DESC
limit 20

''', conn)
```

### **Find Top 10 Customers Returned Frequently.:**

```python
df.columns
```

```python
pd.read_sql('''


SELECT Customer_Name,Count(Customer_Name) FROM store
where Return_Reason != 'Not Returned'
GROUP BY Customer_Name
order by Count(Customer_Name) DESC
limit 10

''', conn)
```

### **Find Top 10  Returned Sub-Categories.:**

```python
pd.read_sql('''


SELECT Sub_Category,Count(Sub_Category) FROM store
where Return_Reason != 'Not Returned'
GROUP BY Sub_Category
order by Count(Sub_Category) DESC
limit 10

''', conn)
```
